## 3. ***Data Wrangling***

### Data Wrangling Code

The top 25 **Countries** which is most affected by terrorism

````python
country_counts = df['country_txt'].value_counts()

top_countries = country_counts.head(25)

print(top_countries)
````

The count of terrorism activities in each **Continents**



````python
continent_counts = df['region_txt'].value_counts()

top_regions = continent_counts.head(20)

print(top_regions)
````

The top 25 **Cities** which is most affected by terrorism

````python
Cities_counts = df['city'].value_counts()

top_cities = Cities_counts.head(25)

print(top_cities)
````

The **type of attacks** that are mostly occurred.

````python

attack_counts1 =df['attacktype1_txt'].value_counts()
attack_counts2 =df['attacktype2_txt'].value_counts()
attack_counts3 =df['attacktype3_txt'].value_counts()
import pandas as pd

# Combine multiple attack type count series into one DataFrame
combined_counts = pd.concat([attack_counts1, attack_counts2, attack_counts3], axis=1)

# Rename the columns in the combined DataFrame
combined_counts.columns = ['attack_type1', 'attack_type2', 'attack_type3']

# Print the combined DataFrame
print(combined_counts)


````

The **Weapon types** used in most of Terrorist attacks.


````python
weapon =df['weaptype1_txt'].value_counts()

print("The weapon types used in Terrorists\n", weapon)
````

The main **Target field** for Terrorist attacks.

````python
target1 =df['targtype1_txt'].value_counts()

print("The main target fields of Terrorists\n", target1)
````

The top 25 primary target organisation proned by terrorist attacks

````python
target_counts = df['corp1'].value_counts()

top_targets = target_counts.head(25)

print(top_targets)
````

The name of the Terrorist **group**  that carried out most of the attacks

````python
group =df['gname'].value_counts().head(25)

print("The main goups which carried most of terrorist attacks\n", group)
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
selected_columns = ['nkill', 'nwound', 'suicide', 'success']
df_selected = df[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selected.corr()

# Create the correlation heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Add a title
plt.title("Correlation Heatmap of Global Terrorism Dataset")

plt.tight_layout()
plt.show()
````

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
numerical_columns = ['nkill', 'nwound', 'nperps', 'nperpcap']

# Drop rows with missing values in the selected columns
df = df.dropna(subset=numerical_columns)

# Create the Pair Plot visualization
plt.figure(figsize=(12, 8))
sns.pairplot(df[numerical_columns], diag_kind='kde')
plt.suptitle("Pair Plot Visualization of Global Terrorism Dataset", y=1.02)

plt.tight_layout()
plt.show()
````
