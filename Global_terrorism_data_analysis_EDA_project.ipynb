#### Chart - 5 - Number of Attacks per Year in Different Region

````python
# Chart - 5 visualization code
grouped_data = df.groupby(['region_txt', 'iyear']).size().reset_index(name='Number_of_Attacks')

# Line plot
plt.figure(figsize=(10, 6))
for continent in df['region_txt'].unique():
    continent_data = grouped_data[grouped_data['region_txt'] == continent]
    plt.plot(continent_data['iyear'], continent_data['Number_of_Attacks'], marker='.', label=continent)

plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title('Number of Attacks per Year in Different Region')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(True)

plt.tight_layout()
plt.show()

````

Chart 14- Some important statistics

In year 2016, more than 75% of terrorist attacks happened within just 10 countries : Iraq, Afghanistan, India, Somalia, Turky, Yemen, Nigeria, Syria, Philippines and Pakistan. Rest of the world combined had less than 25% of total global terrorist attacks.

````python
plt.subplots(figsize=(8,8))
d = pd.DataFrame(df[df['iyear']==2016]['country_txt'].value_counts(sort=True)[:10])
d.reset_index(inplace=True)
d.columns=['country','no_of_attacks']
append = ['Rest of the world',sum(df[df['iyear']==2016]['country_txt'].value_counts(sort=True)) - sum(df[df['iyear']==2016]['country_txt'].value_counts(sort=True)[:10])]
d.loc[len(d)]= append
color_set = ["#e82727","#ff8649","#ffc549","#fced20","#8fe222","#22e28e","#22e2db","#1f85c4","#361599","#d11fce","#00ff99","#fa97fc"]
plt.title("Percentage of attacks in each country")
plt.pie(d['no_of_attacks'],labels=d['country'],colors= color_set,autopct='%.1f%%')
plt.show()
````

#### Chart - 15 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
selected_columns = ['nkill', 'nwound', 'suicide', 'success']
df_selected = df[selected_columns]

# Calculate the correlation matrix
corr_matrix = df_selected.corr()

# Create the correlation heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

# Add a title
plt.title("Correlation Heatmap of Global Terrorism Dataset")

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

To get correlation between entities named 'nkill', 'nwound', 'suicide' and 'success'.

##### 2. What is/are the insight(s) found from the chart?

nkill and nwound are maximum correlated to each other(0.53).

success and suicide are most negatively or minimum correlated to each other (-0.031).
